# Under Sampling

### building a sample dataset  containing similar distribution of noraml transactions and fraudulent transactions

```python
#we are taking a sample equal to the fraudulent data values
legit_sample= legit.sample(n=492)
```

```python
#concatenating the two dataframes
```

```python
new_dataset= pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#now we can do features and target split
```

```python
X= new_dataset.drop(columns= 'Class', axis=1)
Y= new_dataset['Class']
```

```python
Y
```

```python
X
```

```python
#split the data into train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, stratify= Y, random_state= 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## model training

### Logistic Regression

```python
model= LogisticRegression()
```

```python
#training the model with training data
```

```python
model.fit(X_train, Y_train)
```

```python
#model evaluation
```

```python
#accuracy score for training data
X_train_prediction= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
print('training data accuracy:', training_data_accuracy)
```

#### we got a very good accuracy

```python
# accuracy score for test data
X_test_prediction= model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction, Y_test)
print('test data accuracy:', test_data_accuracy)
```

### it is fairly close, hence there is no underfitting or overfitting
